```python
import numpy as np
import pandas as pd
import scipy
import statsmodels.api as sm
import matplotlib as plt
import seaborn as sns
import sklearn
```

```python
data = pd.read_csv("Data\Simple+linear+regression.csv")
```

```python
data.describe()
```

```python
y=data['GPA']
x1=data['SAT']
```

```python
plt.pyplot.scatter(x1,y)
plt.pyplot.show()
```

```python
x= sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
# tomar coheficientes
```
