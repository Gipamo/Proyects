## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import scipy
import statsmodels.api as sm
import matplotlib as plt
import seaborn as sns
import sklearn
```

## Load the data

```python
data = pd.read_csv("Data\_real_estate_price_size.csv")
```

```python
data.head()
```

## Create the regression

### Declare the dependent and the independent variables

```python
y=data['price']
x1=data['size']
```

### Explore the data

```python
plt.pyplot.scatter(x1,y)
plt.pyplot.show()
```

### Regression itself

```python
x= sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()

# Se toman los coheficientes
```

### Plot the regression line on the initial scatter

```python
plt.pyplot.scatter(x1,y)
yhat = x1*223.1787+101900
fig = plt.pyplot.plot(x1,yhat, lw=4, c='orange', label ='regression line')
plt.pyplot.show
```
