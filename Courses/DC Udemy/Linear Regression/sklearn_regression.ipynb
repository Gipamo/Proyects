```python
import numpy as np
import pandas as pd
import scipy
import statsmodels.api as sm
import matplotlib as plt
import seaborn as sns
import sklearn
sns.set()

from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('Data\_regression.csv')
data.head()
```

```python
x=data['SAT']
y=data['GPA']
```

```python
x.shape
y.shape
```

```python
x_matrix = x.values.reshape(-1,1)
```

```python
reg = LinearRegression()
reg.fit(x_matrix,y)
```

```python
# R cuadrado para ver que este bueno
reg.score(x_matrix,y)
```
