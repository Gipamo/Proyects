```python
def anagrams(word, words):
    out=[]
    for x in words:
        if sorted(list(x))== sorted(list(word)):
            out.append(x)
    return out

anagrams('abba', ['aabb', 'abcd', 'bbaa', 'dada']) 
# ['aabb', 'bbaa']
```

```python
def format_duration(seconds):
    out=""
    hour=int(seconds/3600)
    minute=int((seconds%3600)/60)
    second=int((seconds%3600)%60)
    if second != 0:
        if hour > 0 or minute > 0:
            if second == 1:
                out=out+"and "+str(second)+" second"
            else:
                out=out+"and "+str(second)+" seconds"
        else:
            if second == 1:
                out=""+str(second)+" second"+out
            else:
                out=""+str(seconds)+" seconds"+out
    if minute != 0:
        out=" "+out
        if hour > 0 :
            if minute == 1:
                out=", "+str(minute)+" minute"+out
            else:
                out=", "+str(minute)+" minutes"+out
        else:
            if minute == 1:
                out=""+str(minute)+" minute"+out
            else:
                out=""+str(minute)+" minutes"+out
    if hour != 0:
        if hour == 1:
            out=""+str(hour)+" hour"+out
        else:
            out=""+str(hour)+" hours"+out
    return out

        
    







format_duration(1) 
# "1 second")
format_duration(62) 
# "1 minute and 2 seconds")
format_duration(120) 
# "2 minutes")
format_duration(3600)
#  "1 hour")
format_duration(3662) 
# "1 hour, 1 minute and 2 seconds")
```
